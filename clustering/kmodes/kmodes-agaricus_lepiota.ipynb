```python
import numpy as np
import pandas as pd
from kmodes.kmodes import KModes
```

```python
df = pd.read_csv('asset/agaricus-lepiota.csv')
```

```python
df
```

```python
df.describe()
```

```python
dfTarget = df.iloc[:, 0]
dfFeature = df.drop(df.columns[0], axis=1) 
```

```python
dfFeature
dfTarget
```

```python
dfFinal = pd.concat([dfFeature, dfTarget], axis = 1)
dfFinal.head()
```

```python
km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1)
```

```python
clusters = km.fit_predict(dfFinal)
```

```python
kmodes = km.cluster_centroids_
shape = kmodes.shape
```

```python
print(kmodes)
```

```python
for i in range(shape[0]):
    
    print("\ncluster " + str(i) + ": ")
    cent = kmodes[i,:]
    for j in dfFinal.columns[np.nonzero(cent)]:
        print(j)
```
